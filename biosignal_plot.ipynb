# Plotting ECG and EDA

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
path = 'datasets/'
```

### Song and Disturbing video T

```python
with open(path+'noa_song_t.txt') as f:
    contents = f.readlines()
    f.close()
```

```python
# length in minutes at sampling rate 100 Hz
((len(contents)-3)/100)/60
```

```python
# printing labels and first lines
for line in contents[:3]:
    print(line)
```

```python
# create empty lists for ecg and eda
ecg_array = []
eda_array = []
```

```python
# splitting and getting data
for line in contents[3:]:
    list_line = line.split("\t")
    
    ecg_raw_value = int(list_line[5])
    eda_raw_value = int(list_line[6])
    
    ecg_array.append(ecg_raw_value)
    eda_array.append(eda_raw_value)
```

```python
# plot ECG, EDA
ecg_np = np.array(list(enumerate(ecg_array))[60000:64000])
eda_np = np.array(list(enumerate(eda_array))[60000:64000])
```

```python
ecg_np.shape, eda_np.shape
```

```python
plt.figure(figsize=(20,8))
plt.plot(*ecg_np.T)
plt.xlabel("Sample")
plt.ylabel("ECG")
plt.show()
```

```python
plt.figure(figsize=(20,4))
plt.plot(*eda_np.T)
plt.xlabel("Sample")
plt.ylabel("EDA")
plt.show()
```

```python
points_np = np.array(list(zip(ecg_array, eda_array)))
```

```python
points_np.shape
```

```python
plt.figure(figsize=(20,20))
plt.scatter(*points_np.T)
plt.xlabel("ECG")
plt.ylabel("EDA")
plt.show()
```
