```python
import pyedflib
import numpy as np
from matplotlib import pyplot as plt
```

```python
path = 'datasets/'
```

```python
file_name = path + "Part_2_S_Trial1_emotion.bdf"

f = pyedflib.EdfReader(file_name)
n = f.signals_in_file
signal_labels = f.getSignalLabels()

sigbufs = np.zeros((n, f.getNSamples()[0]))

for i in np.arange(n):
    sigbufs[i, :] = f.readSignal(i)
```

```python
sigbufs.min(), sigbufs.max()
```

```python
for i in range(0, sigbufs.shape[0]):
    row = sigbufs[i]
    plt.figure(figsize=(20,2))
    plt.plot(np.array(range(0,len(row))), row)
    plt.grid(True)

    plt.show()
```
